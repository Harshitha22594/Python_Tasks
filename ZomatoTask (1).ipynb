## Handle missing values appropriately


```python
# drop Unnamed: 0 and reset index to 0
df.drop(columns=['Unnamed: 0'], inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# drop url and phone
df.drop(columns=['url', 'phone'], inplace=True)
df
```

```python
# handle missing values in rate column
# Replace 'NEW' and '-' with NaN first
df['rate'].replace(['NEW', '-'], np.nan, inplace=True)

# Ensure column is string type for .str operations and strip whitespace
df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False).str.strip()

# Convert to numeric, coercing any errors to NaN (this handles the string 'nan' correctly)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

# Calculate the mean of the now numeric 'rate' column
avg = df['rate'].mean()

# Fill any remaining NaNs with the calculated average
df['rate'].fillna(avg, inplace=True)

df.isnull().sum()
```

```python
# drop location and rename listed_in(city) as location
df.drop(columns=['location'], inplace=True)
df.rename(columns={'listed_in(city)': 'location'}, inplace=True)
df.head()
```

```python
# handle missing values in cuisines column
df['cuisines'].replace([ np.nan], 0, inplace=True)
mode = mode = df['cuisines'].mode()[0]
df['cuisines'].replace( 0, mode ,inplace=True)
```

```python
# handling missing values in rest_type column
df['rest_type'].replace([ np.nan], 0, inplace=True)
mode = mode = df['rest_type'].mode()[0]
df['rest_type'].replace( 0, mode ,inplace=True)
```

```python
# handling missing values in approx_cost(for two people) column
df['approx_cost(for two people)'].replace([ np.nan], 0, inplace=True)
mode = mode = df['approx_cost(for two people)'].mode()[0]
df['approx_cost(for two people)'].replace( 0, mode ,inplace=True)
```

```python
# handling missing values in dish_liked column
df['dish_liked'].replace([np.nan], 0, inplace=True)
mode = df['dish_liked'].mode()[0]
df['dish_liked'].replace( 0, mode ,inplace=True)
```

```python
# drop address, reviews_list , menu_item column
df.drop(columns=['address'], inplace=True)
df.drop(columns=['reviews_list'], inplace=True)
df.drop(columns=['menu_item'], inplace=True)
```

```python
df['approx_cost(for two people)'].unique()
```

```python
# rename approx_cost(for two people) as cost_for_2 and covert datatype to integer
df.rename(columns={'approx_cost(for two people)': 'cost_for_2'}, inplace=True)
df['cost_for_2'].dtype
# replace , with ""
df['cost_for_2'] = df['cost_for_2'].str.replace(',', '')
df['cost_for_2'] = df['cost_for_2'].astype(int)
```

Perform the following analysis with proper visualizations:

```python
# Top 5 locations with highest number of restaurants
# loaction count using value_count
top_locations = df['location'].value_counts().head(5)
# figure size
plt.figure(figsize=(6, 6))
# bar chart
sns.barplot(x=top_locations.index, y=top_locations.values, palette='viridis')
# title
plt.title('Top 5 locations with highest number of restaurants')
# x axis label
plt.xlabel('Location')
# y axis label
plt.ylabel('Number of Restaurants')
# rotating for x axis
plt.xticks(rotation=45, ha='right')
# good alingment layout
plt.tight_layout()
plt.show()
```

1. Top 5 locations with highest number of restaurants
      
    - we first use value_count of locations and take top 5 of that count.
    - then we use barplot to visualize.
    - we add title, x lable, y lable and xticks to have a proper and understandable visualization.


```python
# Distribution of restaurant ratings
# use histplot with rate and data
sns.histplot(x = df['rate'], data = df, kde = True, color = 'red')
# title
plt.title('Distribution of restaurant ratings')
# x lable
plt.xlabel('Rating')
# y lable
plt.ylabel('Frequency')
# use xticks
plt.xticks(rotation=90)
plt.show()
```

2. Distribution of restaurant ratings

```python
# Percentage of restaurants offering online orders
# store all value count of online order in a value online_orders
online_orders = df['online_order'].value_counts()
# use the online_orders to plot pie chart
plt.pie(online_orders, labels=online_orders.index, autopct='%1.1f%%', shadow = True, explode = (0.2,0))
# use title
plt.title('Percentage of restaurants offering online orders')
plt.show()
```

#Task: 3 Bivariate Analysis
Analyze the relationship between: <br>
• Rating vs Cost <br>
• Rating vs Online Order <br>
• Votes vs Rating <br>
• Use appropriate plots (scatter plot, box plot, bar chart, etc.) <br>

```python
# Rating vs Cost for bivariate analysis
# use scatter plot
plt.scatter(df['rate'],df['cost_for_2'],color = 'red')
# use title
plt.title('Rating vs Cost')
# use x label
plt.xlabel('Rating')
# use y label
plt.ylabel('Cost')
plt.show()
```

As we can see ratings are not given for higher cost item and given more to the lower cost items

```python
# Rating vs Online Order for bivariate analysis using a bar chart of average ratings

# Calculate the average rating for each online_order category
average_ratings = df.groupby('online_order')['rate'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(7, 5))
plt.bar(average_ratings['online_order'], average_ratings['rate'], color=['red', 'lightcoral'])
# use title
plt.title('Average Rating by Online Order Availability')
# use x label
plt.xlabel('Online Order Available')
# use y label
plt.ylabel('Average Rating')
# Set y-axis limits for better visualization
plt.ylim(average_ratings['rate'].min() * 0.9, average_ratings['rate'].max() * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

We can see that average rating in more for resturants which are having online order availability and less for resturants which are not having online order availability

```python
# Votes vs Rating
# creating scatter plot
plt.scatter(df['votes'], df['rate'],color = 'red')
# use title
plt.title('Votes vs Rating')
# use x label
plt.xlabel('Votes')
# use y label
plt.ylabel('Rating')
plt.show()

```

we can see that most people have voted for higher rating
