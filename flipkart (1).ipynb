# Analysing Flipkart sales drop

```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data
customers = pd.read_csv("/content/drive/MyDrive/cap/customers.csv")
orders = pd.read_csv("/content/drive/MyDrive/cap/orders.csv")
products = pd.read_csv("/content/drive/MyDrive/cap/products.csv")
```

```python
#merging the datasets
df = pd.merge(customers,orders,on="customer_id",how="inner")
df = pd.merge(df,products,on="product_id",how="inner")
df
```

# EDA

```python
df.info()
```

```python
#converting object date columns into datetime format
df["signup_date"] =pd.to_datetime(df["signup_date"])
df["order_date"] =pd.to_datetime(df["order_date"])
df.info()
```

```python
#create revenue column
df["revenue"] = df["price"]*df["quantity"]
df.head()
```

```python
#Revenue trend by order date
revenue = df.groupby("order_date")["revenue"].sum()
revenue = pd.DataFrame(revenue).reset_index()
plt.figure(figsize=(15,5))
plt.plot(revenue["order_date"],revenue["revenue"],marker="d",mfc="lavender",mec="red")
plt.xlabel("Order date")
plt.ylabel("Revenue")
plt.title("Revenue by order date")
plt.show()
```

```python
#Revenue trend by city
revenue = df.groupby("city")["revenue"].sum()
revenue = pd.DataFrame(revenue).reset_index()
plt.figure(figsize=(8,5))
plt.plot(revenue["city"],revenue["revenue"],marker="d",mfc="lavender",mec="red")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.title("Revenue by city")
plt.show()
```

```python
#Revenue trend by category
revenue = df.groupby("category")["revenue"].sum()
revenue = pd.DataFrame(revenue).reset_index()
plt.plot(revenue["category"],revenue["revenue"],marker="d",mfc="lavender",mec="red")
plt.xlabel("category")
plt.ylabel("Revenue")
plt.title("Revenue by category")
plt.show()
```

```python
#count customers based on order_id
df.groupby("customer_id")["order_id"].count()
```

```python
#Revenue trend by product_name
revenue = df.groupby("product_name")["revenue"].sum()
revenue = pd.DataFrame(revenue).reset_index()
plt.plot(revenue["product_name"],revenue["revenue"],marker="d",mfc="lavender",mec="red")
plt.xlabel("product_name")
plt.ylabel("Revenue")
plt.title("Revenue by product_name")
plt.show()
```
